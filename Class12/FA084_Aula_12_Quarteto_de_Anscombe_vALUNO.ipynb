# Importação de bibliotecas gerais
* Pandas, NumPy, MatPlotLib e Seaborn

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Conjunto de dados 'Quarteto de Anscombe'

## Estatística básica dos dados

```
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```
# Calcular média de cada x
print('Media x:',round(np.mean(x),2))
print('Media x4:',round(np.mean(x4),2))
```

```
# Calcular desvio padrão de cada x
print('Desvio Padrão x:', round(np.std(x),2))
print('Desvio Padrão x4:', round(np.std(x4),2))
```

```
# Calcular média de cada y
print('Media y1:', round(np.mean(y1),2))
print('Media y2:', round(np.mean(y2),2))
print('Media y3:', round(np.mean(y3),2))
print('Media y4:', round(np.mean(y4),2))
```

```
# Calcular desvio padrão de cada y
print('Desvio Padrão y1:', round(np.std(y1),2))
print('Desvio Padrão y2:', round(np.std(y2),2))
print('Desvio Padrão y3:', round(np.std(y3),2))
print('Desvio Padrão y4:', round(np.std(y4),2))
```

# Regressões lineares para cada grupo ( Retas  y = a * x + b )
## Vamos usar numpy.polyfit(x,y,deg=1)

```
# Reta 1 (x,y1): a1 * x + b1

a1, b1 = np.polyfit(x,y1, deg=1)

x1_fit = np.array(x)
y1_fit = a1*x1_fit + b1
```

```
# Reta 2 (x,y2): a2 * x + b2

a2, b2 = np.polyfit(x,y2, deg=1)

x2_fit = np.array(x)
y2_fit = a2*x2_fit + b2
```

```
# Reta 3 (x,y3): a3 * x + b3

a3, b3 = np.polyfit(x,y3, deg=1)

x3_fit = np.array(x)
y3_fit = a3*x3_fit + b3
```

```
# Reta 4 (x4,y4): a4 * x4 + b4

a4, b4 = np.polyfit(x4,y4, deg=1)

x4_fit = np.array(x4)
y4_fit = a4*x4_fit + b4
```

# Gráficos

```
fig, axs = plt.subplots(2,2, sharex=True, sharey=True, figsize=(12,12));
axs[0, 0].set(xlim=(0, 20), ylim=(2, 14));
axs[0, 0].set(xticks=(0, 10, 20), yticks=(4, 8, 12));

axs[0, 0].plot(x1_fit, y1_fit,'r-', lw=2);
axs[0, 0].scatter(x, y1);

axs[0, 1].plot(x2_fit, y2_fit,'r-', lw=2);
axs[0, 1].scatter(x, y2);

axs[1, 0].plot(x3_fit, y3_fit,'r-', lw=2);
axs[1, 0].scatter(x, y3);

axs[1, 1].plot(x4_fit, y4_fit,'r-', lw=2);
axs[1, 1].scatter(x4, y4);
```
