# "Script" para Sorteio das questões a serem respondidas

```
import random as rnd

# Número de questões (n)
n = 50
# Vetor com n elementos
vetor_n = list(range(1,n+1))

# Solicitar ao usuário o nome completo
n_completo = input('Nome completo: ')

n_completo_lower = n_completo.lower()

vogais = ['a','e','i','o','u']

n_vogais = 0
n_consoantes = 0
n_espacos = 0

for i in n_completo_lower:
    if i in vogais:
        n_vogais += 1
    elif (i ==' '):
        n_espacos += 1 
    else:
        n_consoantes += 1

# Primeira pergunta a responder (R1): Usar seed definido por
# número de letras + número de vogais do seu nome
seed1 = len(n_completo) + n_vogais
rnd.seed(seed1)
R1 = rnd.randint(1, n)

# Segunda pergunta a responder (R2): Usar seed definido por
# número de letras + número de consoantes do seu nome
seed2 = len(n_completo) + n_consoantes
rnd.seed(seed2)
R2 = rnd.randint(1, n)		

# ALTERNATIVA CASO R1 e R2 tenham sido iguais
# Segunda pergunta (alternativa) a responder (R2): 
# Usar seed definido por número de letras dividido por número de vogais e consoantes do seu nome
if R1 == R2:
    seed2_alt = len(n_completo) / (n_consoantes + n_vogais)
    rnd.seed(seed2_alt)
    R2 = rnd.randint(1, n)

print('Primeira pergunta:', R1)
print('Segunda pergunta:', R2)
```
