# Regressão

## Regressão Linear

```
# Importar bibliotecas
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
# Criar um exemplo com duas variáveis: x e y
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```
# Criar um regressor (lrm: linear regression model)
lrm = LinearRegression()
```

```
# Fazer ajuste da curva utilizando o regressor lrm e os vetores x e y
lr_modelo = lrm.fit(x,y)

# ALTERNATIVA
# Fazer tudo em uma única linha
# lr_modelo = LinearRegression().fit(x,y)
```

```
# Calcular o coeficiente de determinação do modelo (R quadrado)
r_sq = lr_modelo.score(x,y)
r_sq
```

```
# Obter os valores dos coeficientes b0 e b1
b0 = lr_modelo.intercept_  # (valor único)
b1 = lr_modelo.coef_       # (vetor)
```

```
b1
```

```
# Predições
y_pred = lr_modelo.predict(x)

# OBS: Os valores de y_pred obtidos por este comando necessariamente 
#      devem ser iguais ao obtidos por "lr_modelo.intercept_ + lr_modelo.coef_ * x"
```

## Regressão Linear Múltipla

```
# Importar bibliotecas
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
# Criar conjuntos de dados: x (matriz) e y
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
```

```
# Criar regressor mlrm (Multiple Linear Regression Model)
mlrm = LinearRegression()
```

```
# Fazer ajuste da curva utilizando o regressor mlrm e os vetores x e y
mlr_modelo = mlrm.fit(x,y)
```

```
# Calcular o coeficiente de determinação do modelo (R quadrado)
r_sq = mlr_modelo.score(x,y)
r_sq
```

```
# Obter os valores dos coeficientes b0, b1 e b2
b0 = mlr_modelo.intercept_
b1b2 = mlr_modelo.coef_
```

```
b1b2
```

```
# Predições
y_pred = mlr_modelo.predict(x)
```

```
y_pred
```

## Regressão Polinomial

```
# Importar bibliotecas
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```
# Criar um exemplo com duas variáveis: x e y
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])
```

```
# ATENÇÃO!!! NOVO PASSO!!!  NECESSÁRIO PARA REGRESSÃO POLINOMIAL!!!
# Transformar os dados de entrada conforme o grau do polinômio a ser utilizado
transformer = PolynomialFeatures(degree=2, include_bias=False)
```

```
# Utilizando o vetor x, fazer o ajuste do tranformador definido anteriormente
transformer.fit(x)
```

```
# Criar matriz com valores ajustados com o transformador
x_ = transformer.transform(x)
```

```
# ALTERNATIVA
# Fazer tudo em uma única linha
# x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
```

```
# x_ retorna uma matriz com os valores de entrada transformados.
# Neste exemplo, a primeira coluna com os valores originais e a segunda, 
# com os quadrados desses valores
# Imprimir x_
print(x_)
```

```
# Fazer ajuste da curva utilizando os vetores x e y
prm = LinearRegression().fit(x_,y)   # prm: Polynomial Regression Model
```

```
# Calcular o coeficiente de determinação do modelo (R quadrado)
r_sq = prm.score(x_,y)
r_sq
```

```
# Obter os valores dos coeficientes b0, b1 e b2
b0 = prm.intercept_
b1b2 = prm.coef_
```

```
b1b2
```

```
# OBS IMPORTANTE!!! Se incluir bias=True matriz modificada vai incluir 
# intercept = 1 para todos os registros

# Nesse caso, quando for fazer a regressão, fit_intercept deve ser False
# LinearRegression(fit_intercept=False).fit(x_, y)

# No resultado, 'intercept' vai ser 0.0, mas nos coeficientes, vai aparecer
# o primeiro valor como sendo o valor b0 de um modelo y = b0 + b1 x + b2 x^2
```

```
# Predições
y_pred = prm.predict(x_)
```

```
# Criar matriz com valores ajustados com o transformador (tudo em uma única linha)
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
```

```
# Ajustar o modelo utilizando a matriz transformada e o vetor y
model = LinearRegression().fit(x_, y)
```

```
# Obter os valores dos coeficientes b(i) - Comando em uma única linha
# OBS: Neste exemplo, vão existir seis coeficientes: 
# b0, b1, b2, b3, b4 e b5
# f(x1,x2) = b0 + b1 x1 + b2 x2 + b3 x1^2 + b4 x1 x2 + b5 x2^2
intercept, coefficients = model.intercept_, model.coef_   
```

```
# Predições utilizando a matriz transformada
y_pred = model.predict(x_)
```
