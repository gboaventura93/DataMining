## Função relplot(), do Seaborn

### O método relplot() combina dois tipos de gráficos: scatter plot e line plot

### Útil quando se pretende verificar relações existentes entre duas variáveis

```
# Faça uma consulta ao 'Help' para ver os principais parâmetros
?sns.relplot
```

```
# Gráfico scatter plot com 'sepal_length' no eixo x e 'sepal_width' no eixo y
# Default: kind = 'scatter'
sns.relplot(x='sepal.length', y='sepal.width', data=df);
```

```
# Gráfico scatter plot com 'sepal_length' no eixo x e 'sepal_width' no eixo y
# Experimentar incluir kind = 'line'
sns.relplot(x='sepal.length', y='sepal.width', kind = 'line', data=df);
```

```
# Gráfico scatter plot com 'petal_length' no eixo x e 'petal_width' no eixo y
sns.relplot(x='petal.length', y='petal.width', data=df);
```

```
# Incluir nos gráficos acima cores diferentes para espécies diferentes (hue)
# 'sepal_length' no eixo x e 'sepal_width' no eixo y
sns.relplot(x='sepal.length', y='sepal.width', hue='species', data=df);
```

```
# Incluir nos gráficos acima cores diferentes para espécies diferentes
# 'petal_length' no eixo x e 'petal_width' no eixo y
```

```
# Nos gráficos acima, todos os pontos tinham o mesmo formato.
# Pode-se alterar os gráficos acima de forma que os pontos terão formatos 
# diferentes para cada espécie (style), além das cores diferenciadas
# 'petal_length' no eixo x e 'petal_width' no eixo y
sns.relplot(x='sepal.length', y='sepal.width', hue='species', style='species', data=df);
```

```
# Pode-se também incluir o parâmetro 'col'.
# Quando esse parâmetro é utilizado, o gráfico é feito em colunas, 
# sendo que em cada coluna será feito o gráfico para cada valor possível 
# daquele atributo.

# Deve-se especificar qual o atributo para 'col'. Por exemplo: col='sex',
# se fosse o caso de haver um atributo 'sex' nesse conjunto de dados.

# Além do parâmetro 'col', pode-se também incluir o parâmetro 'row'.
# É similar ao 'col', mas o gráfico vai fazer a separação por linhas.

# Se 'col' e 'row' forem utilizados em conjunto, o gráfico vai ser 
# apresentado na forma de matriz, com linhas e colunas.
```

## Especificando a tonalidade (mais clara ou mais escura) e a espessura (mais fina ou mais grossa) da linha de regressão, bem como o tamanho dos pontos

```
# scatter_kws={'s': valor_tamanho_ponto, 'alpha': valor_tonalidade}
# line_kws={'lw': valor_espessura_linha}
sns.lmplot(x='sepal.length', 
           y='sepal.width', 
           hue='species', 
           data=df);
```

## Especificando o estilo das linhas para cada função plotada: linha cheia, pontilhada, etc.
## Usar linestyle = 'estilo_linha',
## Estilos de linha podem ser: 'solid', 'dashed', 'dashdot', 'dotted'

```
# Construir quatro gráficos sin(x), com defasagem entre eles de pi/4 no eixo x,
# cada um com um estilo de linha.

plt.plot(x, np.sin(x - 0), linestyle = 'solid')
plt.plot(x, np.sin(x - np.pi/4), linestyle = 'dashed')
plt.plot(x, np.sin(x - np.pi/2), linestyle = 'dashdot')
plt.plot(x, np.sin(x - 3*np.pi/4), linestyle = 'dotted')
```

## Combinando em um único comando o estilo das linhas e a cor para cada função plotada
*   Tipo de linha
> **'-'**  solid  
> **'--'** dashed  
> **'-.'** dashdot  
> **':'**  dotted

*   Cor (primeira letra)
> **'g'** green  
> **'r'** red  
> **'b'** blue  
> **'m'** magenta  
> **'y'** yellow  
> **'c'** cyan

```
# Construir quatro gráficos sin(x), com defasagem entre eles de pi/4 no eixo x,
# cada um com um estilo de linha e uma cor, combinando a linha e a cor em um 
# único comando.
plt.plot(x, np.sin(x - 0), '-g')           # solid green
plt.plot(x, np.sin(x - np.pi/4), '--r')     # dashed red
plt.plot(x, np.sin(x - np.pi/2), '-.b')     # dashdot blue
plt.plot(x, np.sin(x - 3*np.pi/4), ':m')   # dotted magenta
```

## Extrair ano, mês e dia de um atributo no formato 'dd/mm/yy'

```
# Importar pandas, ler o arquivo e apresentar as suas primeiras linhas
import pandas as pd
df = pd.read_csv('/content/CidadeA_15-01-2001_a_15-02-2001.csv')
df.head()
```

```
# Converter o formato da data dd/mes/ano (dd/mm/yy) para ano-mes-dia (yyyy/mm/dd)
# Exemplo:  05/03/01  (5 de Março de 2001)==>   2001-03-05

# Determinar o número de linhas do arquivo visando o loop
# Pseudo código
#   1. Calcular número de linhas do dataset
       nlinhas = df.shape[0]
#   2. Loop por todas as linhas
#   2.1. Em cada linha, identificar o conteúdo da coluna 'Data'
#   2.2. Separar cada parte (dia, mes, ano) e reorganizar conforme desejado
#        alocando nesse novo formato à própria coluna 'Data'
#        Exemplo: 05/03/01
#                 ^^^^^^^^
#                 01234567  (indice de cada caracter da data)
#        a) Ano: Acrescentar "20" às duas últimas (no exemplo, transformar '01' em '2001')
#        b) Mes: Quarto (indice 3) e quinto (indice 4) caracter
#        c) Dia: Primeiro (indice 0) e segundo (indice 1) caracter 
#        d) Acrescenter um hifen ('-') entre Ano, Mes e Dia (Ano-Mes-Dia)
#        e) Alocar esse novo formato à coluna 'Data' na linha que está sendo trabalhada
for i in range(nlinhas):
    df_Item=df.loc[i,'Data']
    df.loc[i,'Data']="20"+df_Item[6:]+"-"+df_Item[3:5]+'-'+df_Item[:2]

df['Data'].head()
```
